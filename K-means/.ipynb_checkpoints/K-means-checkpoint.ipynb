```python
#import section

import numpy as np
import pandas as pd
import matplotlib.pyplot as mat

from sklearn.cluster import KMeans
```

```python
#file declaration and locking series

data = pd.read_csv('..//src//main.csv')
x = data.iloc[:, [2, 3]].values
```

```python
#recipe for K-Means clustering and plotting a graph to find the elbow i.e WCSS (the sum of square of each particle in a cluster)

li= []

def kmeans(i):
    k=KMeans(n_clusters = i, random_state=42)
    k.fit(x)


for i in range(1, 11):
    kmeans(i)
    li.append(k.inertia_)

mat.plot(range(1, 11), li)
mat.title('K-means elbow graph')
mat.xlabel('clusters')
mat.ylabel('list')

mat.show
```

```python
#init K-Means and computing clusters 

k1 = KMeans(n_clusters = 3, init= 'k-means++', random_state = 42)
y_pred= k1.fit_predict(x)
```

```python
#plotting a scatter plot for visualization

mat.scatter(x[y_pred == 0, 0], x[y_pred == 0, 1], s = 100, c = 'blue', label = 'Cluster 1') #for first cluster  
mat.scatter(x[y_pred == 1, 0], x[y_pred == 1, 1], s = 100, c = 'green', label = 'Cluster 2') #for second cluster  
mat.scatter(x[y_pred== 2, 0], x[y_pred == 2, 1], s = 100, c = 'red', label = 'Cluster 3') #for third cluster  

mat.scatter(k1.cluster_centers_[:, 0], k1.cluster_centers_[:, 1], s = 300, c = '#ffbb00', label = 'Centroid')   
mat.title('Clusters of customers')  
mat.xlabel('Annual Income (k$)')  
mat.ylabel('Spending Score (1-100)')  
mat.legend()  
mat.show()  
```
